```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Optional: show all columns
pd.set_option('display.max_columns', None)

# Set plot style
sns.set(style='whitegrid', palette='muted', font_scale=1.1)

# === 1. Basic Info ===
print("Shape:", cap_inv.shape)
print("\nColumns:", cap_inv.columns.tolist())
cap_inv.info()
print(cap_inv.describe(include='all'))

# === 2. Missing Values ===
missing = cap_inv.isnull().sum()
missing_pct = 100 * missing / len(cap_inv)
missing_df = pd.DataFrame({'Missing Values': missing, 'Percent': missing_pct})
print("\nMissing Value Summary:\n", missing_df[missing_df['Missing Values'] > 0].sort_values(by='Percent', ascending=False))

# === 3. Numerical Distribution ===
numerical_cols = cap_inv.select_dtypes(include=np.number).columns

cap_inv[numerical_cols].hist(bins=30, figsize=(15, 12))
plt.suptitle("Numerical Features Distributions", fontsize=16)
plt.tight_layout()
plt.show()

# === 4. Correlation Heatmap ===
plt.figure(figsize=(12, 8))
sns.heatmap(cap_inv[numerical_cols].corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Matrix")
plt.show()

# === 5. Categorical Features ===
categorical_cols = ['Investment_Type', 'Protection_Type']

for col in categorical_cols:
    plt.figure(figsize=(8, 4))
    sns.countplot(data=cap_inv, x=col, order=cap_inv[col].value_counts().index)
    plt.title(f'Distribution of {col}')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

# === 6. Boxplots for Cost Features by Type ===
cost_vars = ['Total_Cost_USD', 'Monthly_Cost_USD', 'Cost_Benefit_Ratio']

for var in cost_vars:
    plt.figure(figsize=(10, 4))
    sns.boxplot(data=cap_inv, x='Investment_Type', y=var)
    plt.xticks(rotation=45)
    plt.title(f'{var} by Investment Type')
    plt.tight_layout()
    plt.show()

# === 7. Log-Transform Skewed Variables ===
for col in cost_vars:
    cap_inv[f'log_{col}'] = np.log1p(cap_inv[col])

sns.pairplot(cap_inv[[f'log_{col}' for col in cost_vars] + ['Cost_Benefit_Ratio']], diag_kind='kde')
plt.show()

# === 8. Monthly Cost Distribution (List Column) ===
cap_inv['Monthly_Cost_Distribution'] = cap_inv['Monthly_Cost_Distribution'].apply(eval)
cap_inv['Cost_Months'] = cap_inv['Monthly_Cost_Distribution'].apply(len)
print("Length of Monthly Cost Distributions:\n", cap_inv['Cost_Months'].value_counts())

plt.figure(figsize=(10, 5))
for i in range(min(5, len(cap_inv))):
    plt.plot(cap_inv.loc[i, 'Monthly_Cost_Distribution'], label=f'Project {cap_inv.loc[i, "Project_ID"]}')
plt.title("Monthly Cost Distributions (First 5 Projects)")
plt.xlabel("Month")
plt.ylabel("Cost")
plt.legend()
plt.show()

# === 9. New Features and Ratios ===
cap_inv['Cost_per_Meter'] = cap_inv['Total_Cost_USD'] / cap_inv['Area_Effect_Meters']
cap_inv['Cost_per_Life_Saved'] = cap_inv['Total_Cost_USD'] / (cap_inv['Lives_Saved_Per_Event'] + 1e-6)

sns.scatterplot(data=cap_inv, x='Cost_per_Meter', y='Cost_Benefit_Ratio', hue='Investment_Type')
plt.title("Cost per Meter vs Cost Benefit Ratio")
plt.xscale('log')
plt.yscale('log')
plt.show()

# === 10. Text Description Feature ===
cap_inv['Description_Length'] = cap_inv['Description'].apply(lambda x: len(str(x).split()))

sns.histplot(cap_inv['Description_Length'], bins=20, kde=True)
plt.title("Distribution of Description Length")
plt.show()

# === 11. Top ROI Projects ===
top_roi = cap_inv.sort_values(by='Cost_Benefit_Ratio', ascending=False).head(5)
print("\nTop 5 Projects by Cost Benefit Ratio:\n", top_roi[['Project_ID', 'Investment_Type', 'Cost_Benefit_Ratio', 'Total_Cost_USD']])
```
